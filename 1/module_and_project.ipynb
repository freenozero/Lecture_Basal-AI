이번 강의에서는 파이썬 프로젝트의 기본이 되는 **모듈**과 **패키지**, 그리고 **프로젝트**의 개념에 대해서 배웁니다.

우리는 앞서 파이썬에서 제공하는 여러가지 모듈들을 사용했습니다. 이러한 모듈과 패키지를 구성하고, 실제로 다른 개발자가 만든 모듈을 사용하는 방법까지 이 챕터에서 다루게 됩니다.

내용은 많이 어렵지는 않지만, 알면 여러모로 도움되는 내용이니 많은 연습을 하고 넘어가면 좋겠습니다.

OOP(객체) < 모듈 < 프로젝트(패키지가 공개된 상태)  (<는 부분집합)<br>

### module overview
- 프로그램에서는 작은 프로그램 조각들, 모듈을 모아서 하나의 큰 프로그램을 개발한다.
- 프로그램을 모듈화하면 다른 프로그램이 사용하기가 쉽다. 예) 카카오 게임을 위한 카카오통 접속 모듈<br>
- Built-in Moudle인 Random을 사용할 수 있다.

### 패키지
- 하나의 대형 프로젝트를 만드는 코드의 묶음
- 다양한 모듈들의 합, 폴더로 연결됨
- __init__, __main__ 등 키워드 파일명이 사용됨
- 다양한 오픈 소스들이 모두 패키지로 관리됨

# Modul 만들기
1. 파이썬의 Moudle == py 파일을 의미한다.
2. 같은 폴더에 Module에 해당하는 .py 파일과 사용하는 .py을 저장한 후
3. import 문을 사용해서 module을 호출한다.

## Namespace
- 모듈을 호출할 때 범위를 정하는 방법
- 모듈 안에는 함수와 클래스 등이 존재 가능
- 필요한 내용만을 골라서 호출할 수 있음
- from과 import 키워드를 사용
```Python
#Alias 설정하기 - 모듈명을 별칭으로 써서..
import fah_converter as fah
print(fah.convert_c_to_f(41.6))
#모듈에서 특정 함수 또는 클래스만 호출..
from fah_converter import convert_c_to_f
print(convert_c_to_f(41.6))
#모듈에서 모든 함수 또는 클래스를 호출.
from fah_converter import *
print(convert_c_to_f(41.6))#전체호출


## Built-in Module
- 파이썬이 기본 제공하는 라이브러리로 import로 활용 가능 예)random, time, urllib.request...


# 패키지 만들기
1. 기능들을 세부적으로 나눠 폴더로 만듦
2. 각 폳러별로 필요한 모듈을 구현함
3. 폴더별로 __init__.py 구성하기
    - 현재 폴더가 패키지임을 알리는 초기화 스크립트
    - 없을 경우 패키지로 간주하지 않음(3.3+ 부터는 아님)
    - 하위 폴더와 py파일(모듈)을 모두 포함함
    - import와 __all__ keyword 사용
4. __main__.py 파일 만들기

> **Game 안에 image, sound, stage 폴더와 __init__.py, __main__.py<br>**
> image 폴더 안에 __init__.py, character.py, object_type.py<br>
> sound 폴더 안에 __init__.py, echo.py, bgm.py<br>
> stage 폴더 안에 __init__.py, main.py, sub.py<br>

> ### **Game** 안에 __init__.py
> __all__ = ["image", "sound", "stage"]<br>
> from . import image <br>
> from . import sound <br>
> from . import stage <br>

> ### **image** 안에 __init__.py
> __all__ = ["character", "object_type"]<br>
> from . import character <br>
> from . import object_type <br>

> ### **sound** 안에 __init__.py
> __all__ = ["echo", "bgm"]<br>
> from . import echo <br>
> from . import bgm <br>

> ### **stage** 안에 __init__.py
> __all__ = ["main", "sub]<br>
> from . import main <br>
> from . import sub <br>


<br>

> ### 폴더자체를 실행시킬 때
> ### **Game** 안에 __main__.py
 
>```Python
from stage.main import game_start
from stage.sub import set_stage_level
from image.character import show_character
from sound.bgm import bgm_play
if __name__ == '__main__':
    game_start()
    set_stage_level(5)
    bgm_play(10)
    show_character()
> ```

### package namespace
- package 내에서 다른 폴더의 모듈을 부를 때 상대 참조로 호출하는 방법

```Python
from```Python game.graphic.render import render_test() #절대참조
from .render import render_test() #현재 디렉토리 기준

from ..sound_echo import echo_test() #부모 디렉토리 기준 
#EX) bgm 모듈에서 image에 있는 모듈을 부른다면
from ..image import character()
```
